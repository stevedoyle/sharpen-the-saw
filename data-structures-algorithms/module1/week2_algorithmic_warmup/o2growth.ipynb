```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = np.linspace(1, 100)

plt.plot(n, np.emath.logn(4, n), label="log4(n)")
plt.plot(n, n ** .5, label="n^.5")
plt.plot(n, n, label="n")
plt.plot(n, n*np.log2(n), label="nlog2(n)")
plt.plot(n, n ** 2, label="n^2")

#plt.plot(n, 5 ** (np.log2(n)), label="5^(log2(n))")
#plt.plot(n, 3 ** n, label="3n")
#plt.plot(n, 4 ** (n), label="2^(2n)")

plt.legend(loc='upper left')
plt.show()
```
